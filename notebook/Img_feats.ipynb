```python
train_feats_list = ['../input/feats_0.pkl',
 '../input/feats_1.pkl',
 '../input/feats_2.pkl',
 '../input/feats_3.pkl',
 '../input/feats_4.pkl']
```

```python
train_feats = pd.DataFrame()
for pkl in train_feats_list:
    df = pd.read_pickle(pkl)
    train_feats = train_feats.append(df)

train_feats.info()
```

```python
# rescale img feature
train_feats.blurrness = train_feats.blurrness.apply(lambda x: np.log(x+1))
train_feats.blurrness.describe()
```

```python
rescale_feats = [
 'blurrness',
 'dullness',
 'height',
 'size',
 'whiteness',
 'width']
```

```python
for i in rescale_feats:
    train_feats[i] = train_feats[i].apply(lambda x: np.log(x+1))
```
